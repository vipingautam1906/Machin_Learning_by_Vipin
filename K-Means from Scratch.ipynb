```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import multivariate_normal
```

```python
MU1 = [-1,2]
MU2 = [3,3]
MU3 = [7,4]
cov1 = [[3, 2], [2, 3]]                                      
cov2 = [[3, -2], [-2, 3]]
cov3 = [[2.5, 1.5], [1.5, 2.5]]


X1 = np.linspace(-5,12,100)  
X2 = np.linspace(-5,12,100)
X, Y = np.meshgrid(X1,X2) 

gause_1 = multivariate_normal(MU1, cov1)  
gause_2 = multivariate_normal(MU2, cov2)
gause_3 = multivariate_normal(MU3, cov3)

ar = np.empty(X.shape + (2,))                
ar[:, :, 0] = X 
ar[:, :, 1] = Y   
                                                          
size = plt.figure()
size.set_figwidth(10)
size.set_figheight(10)
plt.contour(X, Y, gause_1.pdf(ar), colors="r" ) 
plt.contour(X, Y, gause_2.pdf(ar), colors="g" )
plt.contour(X, Y, gause_3.pdf(ar), colors="b" )
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
PI1 = 0.5
PI2 = 0.3
PI3 = 0.2
size = plt.figure()
size.set_figwidth(10)
size.set_figheight(10)
# Combined Plot
plt.contour(X, Y, PI1*gause_1.pdf(ar)+PI2*gause_2.pdf(ar)+PI3*gause_3.pdf(ar), colors="b" )
```

```python
ax = plt.axes(projection='3d')
ar = np.empty(X.shape + (2,))                
ar = np.empty(X.shape + (2,))                
ar[:, :, 0] = X 
ar[:, :, 1] = Y 


ax.plot_surface(X, Y, PI1*gause_1.pdf(ar)+PI2*gause_2.pdf(ar)+PI3*gause_3.pdf(ar))

plt.show()
```
